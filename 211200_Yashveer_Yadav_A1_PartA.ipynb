```python
# Importing Libraries : Random, Numpy, Matplotlib.pyplot
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This is the Montecarlo fuction which will do Random Gaussian Sampling!
# And Return the Standard Deviation and Mean of the sample output.
def montecarlo_normal(mu,sigma, num_samples):
    samples= np.random.normal(mu,sigma, num_samples)

    mean_estimate=np.mean(samples)
    std_dev_estimate=np.std(samples)

    return mean_estimate, std_dev_estimate
```

```python
#Main Function
if __name__ == "__main__":
    #Mean of the Gaussian Distribution
    mu=0
    # Deviation of the Gaussian Random variable
    sigma= 0.1
    # Given in Question 
    epsilon = 0.05
    # Initialising the variable for the number of samples we want.
    num_samples=0
    # Initiallising list for X-coordinate and Y-coordinate which will be used to plot the graph
    x_coor=[]
    y_coor=[]
    # Arrays of Some pre-Computation.
    # ith entry of y represent how many times does sample mean > epsilon when i+1 samples are taken(i.e. n=i+1)
    y=[0]*50
    # Selecting the number of samples we want to take.
    for num_samples in range(1,50):
        # Doing 500 experiments.
        for i in range(0,500):
            mean_estimate,std_dev_estimate=montecarlo_normal(mu,sigma,num_samples)
            if(mean_estimate>=epsilon):
                y[num_samples-1]=y[num_samples-1]+1

    for i in range(0,50):
            #Storing the value of n
            x_coor.append(i+1)
            #Storing the emperical estimate of probability of mean estimate greater than epsilon.
            y_coor.append(y[i]/500)
    
```

```python
# Plotting the exponential Curve
a=np.arange(1,50)
b=np.exp(-0.125*a)
plt.plot(a,b,label="Exponential Curve")
# Plotting the Data of P(x>=epsilon) vs N
plt.scatter(x_coor,y_coor,color='red')
# Labelling the Graph
plt.xlabel("Number of Samples---------->")
plt.ylabel("Emperical Estimate of Probability---------->")
plt.title("Emperical Estimate vs Number of Samples")
```
